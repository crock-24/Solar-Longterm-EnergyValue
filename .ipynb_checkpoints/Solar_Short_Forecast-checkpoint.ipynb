```python
import pandas as pd
```

```python
## Loading in a sample of real time MI solar farm production data
# data from https://www.nrel.gov/grid/solar-power-data
solar_sample = pd.read_csv('Actual_42.55_-85.35_2006_DPV_15MW_5_Min.csv')

## grabbing average daily solar output data from current season
solar_sample['LocalTime'] = pd.to_datetime(solar_sample['LocalTime'], format='%m/%d/%y %H:%M')
solar_sample['LocalTime'] = solar_sample['LocalTime'].dt.date
daily_avg_solar = solar_sample.groupby("LocalTime")[["Power(MW)"]].mean().reset_index()
```

```python
train_solar = daily_avg_solar.loc[0:340,:]
test_solar = daily_avg_solar.loc[340:365,:]
```

```python
## grabbing historical cloud cover data, will be used to adjust solar models to account for any abnormal cloud coverage over the sample period
from datetime import datetime, timedelta
from dateutil.relativedelta import relativedelta
import requests

# Define parameters
latitude = '42.55'
longitude = '85.35'

# number of years to collect historical data from in the same time period as solar sample
start = daily_avg_solar.loc[0,'LocalTime']
end = daily_avg_solar.loc[len(daily_avg_solar)-1,'LocalTime']

daily_cloud = pd.DataFrame(columns = ['Date', 'cloud_cover'])

timezone = "America/Detroit"
    
# Construct the API URL
url = (
    f"https://archive-api.open-meteo.com/v1/archive?"
    f"latitude={latitude}&longitude={longitude}"
    f"&start_date={start}&end_date={end}"
    f"&hourly=cloudcover&timezone={timezone}"
)

# Make the API request
response = requests.get(url)
data = response.json()

# Convert to DataFrame
df = pd.DataFrame({
    'time': data['hourly']['time'],
    'cloud_cover': data['hourly']['cloudcover']
})

df['time'] = pd.to_datetime(df['time'])

# Group by date and calculate average cloud cover
df["Date"] = df["time"].dt.date
daily_avg_cloud = df.groupby("Date")[["cloud_cover"]].mean().reset_index()
```

```python
## converting into cloud cover over prediction and cloud cover forecast
train_cloud = daily_avg_cloud.loc[0:340,:]
test_cloud = daily_avg_cloud.loc[340:365,:]
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.deterministic import CalendarFourier, DeterministicProcess
import pandas as pd
import matplotlib.pyplot as plt

# First, check if 'date' is already the index or if it's a column
train_solar = train_solar.set_index('LocalTime')

# Make sure the index is datetime and set frequency
train_solar.index = pd.to_datetime(train_solar.index)
train_solar = train_solar.asfreq('D')  # Set daily frequency

# Create Fourier seasonal terms
fourier = CalendarFourier(freq='A', order=4)  # annual seasonality, 3 harmonics
dp = DeterministicProcess(index=train_solar.index, constant=True,
                          order=1,
                          seasonal=False,
                          additional_terms=[fourier],
                          drop=True)

X = dp.in_sample()
model = SARIMAX(train_solar['Power(MW)'], order=(1, 0, 1), exog=X)
result = model.fit()
print(result.summary())

# Forecast 7 days ahead
n_forecast = 7
X_future = dp.out_of_sample(steps=n_forecast)

# Using 'result' variable for consistency
forecast = result.get_forecast(steps=n_forecast, exog=X_future)
forecast_ci = forecast.conf_int()
```

```python
# Plot forecast
plt.figure(figsize=(15,5))
train_solar['Power(MW)'].plot(label='Observed')
forecast.predicted_mean.plot(label='Forecast')
plt.fill_between(forecast_ci.index, forecast_ci.iloc[:,0], forecast_ci.iloc[:,1], color='lightgrey', alpha=0.5)
plt.legend()
plt.title('Forecast with ARIMA + Fourier Terms')
plt.show()
```
